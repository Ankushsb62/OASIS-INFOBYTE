```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv('AB_NYC_2019.csv')
data.head()
```

```python
# Drop rows with missing values
data = data.dropna()
print(f'Number of rows after dropping missing values: {data.shape[0]}')
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
print(f'Number of rows after removing duplicates: {data.shape[0]}')
```

```python
# Convert categorical variables to numerical using one-hot encoding
data = pd.get_dummies(data, columns=['room_type', 'neighbourhood_group', 'neighbourhood'])

# Print the column names after one-hot encoding
print(data.columns)
```

```python
# Detect outliers using z-score
from scipy import stats
z_scores = np.abs(stats.zscore(data[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']]))
outliers = (z_scores > 3).any(axis=1)
data = data[~outliers]
print(f'Number of rows after removing outliers: {data.shape[0]}')
```

```python
# Save the cleaned dataset
data.to_csv('cleaned_dataset.csv', index=False)
print('Cleaned dataset saved as "cleaned_AB_NYC_2019.csv"')
```
